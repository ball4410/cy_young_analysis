```python
# Dependencies
from bs4 import BeautifulSoup
import requests


url = "https://www.espn.com/mlb/features/cyyoung"
response = requests.get(url)


soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find('table', class_="boxed")
table_rows = results.find_all("tr")

# table_headers = results.find_all("td", class_="banner")
# table_years = results.find_all("td", class_="dataColBlueC")
# headers = []

# #Loop through returned results
# for header in table_headers:
#     # Error handling
#     try:
#         # Identify and return title of listing
#         title = header.text
#         headers.append(title)

#         # Print results only if title, price, and link are available and price and link
#         if (title):
#             print('-------------')
#             print(title)
#     except AttributeError as e:
#         print(e)

for row in table_rows:
    table_headers = row.find_all("td", class_="banner")
    
    for header in table_headers: 
    
        # Error handling
        try:
            # Identify and return title of listing
            title = header.text
            headers.append(title)

            # Print results only if title, price, and link are available and price and link
            if (title):
                print('-------------')
                print(title)
        except AttributeError as e:
            print(e)
```

```python
import pandas as pd

tables = pd.read_html(requests.get(url).text)
tables
```

```python
df = tables[0]
df.head()

al_df = df.iloc[1:12:]

nl_df = df.iloc[13:]

al_df
```

```python
new_header = al_df.iloc[0] #grab the first row for the header
al_df = al_df[1:] #take the data less the header row
al_df.columns = new_header

al_df
```

```python
new_header = nl_df.iloc[0] #grab the first row for the header
nl_df = nl_df[1:] #take the data less the header row
nl_df.columns = new_header

nl_df
```

```python
frames = [al_df, nl_df]

po = pd.concat(frames)
po_df = po.reset_index(drop = True)

po_df
```

```python
po_df['W-L'].str.split("-", expand=True)
po_df[['W','L']] = po_df['W-L'].str.split("-", expand=True)
po_df.set_index(po_df["RK"])

po_df = po_df[["PLAYER", "W", "K", "ER", "IP", "L", "SHO", "SV"]]
```

```python
po_df.to_csv("player_options.csv")
```
